```python
# Load the dataset
# accessing data using xlrd library 
xldata = xlrd.open_workbook('ccc.xls')
# accessing sheet 1
data_sheet = xldata.sheet_by_index(0)
#coloumn 2 Limit bal
limit_bal=data_sheet.col_values(1)[1:]
# converting to numpy
np_lim_bal=np.array(limit_bal)
print(len(np_lim_bal))
```

```python
# Find the average, mean, median and standard deviation of one numerical attribute
avg=np.average(np_lim_bal)
print(f"Average of Rank: {avg:.2f}")

mn=np.mean(np_lim_bal)
print(f"Mean of Rank: {mn:.2f}")

med=np.median(np_lim_bal)
print(f"Median of Rank: {med:.2f}")

stndrd_dev=np.std(np_lim_bal)
print(f"Standard Deviation of Rank: {stndrd_dev:.2f}")
```

```python
sex=data_sheet.col_values(2)[1:]
np_sex=np.array(sex)
edu=data_sheet.col_values(3)[1:]
np_edu=np.array(edu)

# create a 2D array using list comprehension 
arr = [[np_sex[i], np_edu[i]] for i in range(len(np_sex))]

# print the resulting array
print(arr)
print(len(arr))
print(1089**2)
```

```python
# p-m/p
p=2
m=0
dsimm_arr=[]
for i in range(len(arr)-1, 0, -1):
    for j in range(i-1,-1,-1):
        m=0
        if(arr[i][0]==arr[j][0] and arr[i][1]==arr[j][1]):
            m=2
        elif(arr[i][0]==arr[j][0] or arr[i][1]==arr[j][1]):
            m=1
        dsimm_arr.append((p-m)/p)

print(f"Number of disimilarity relationships: {len(dsimm_arr)}")
#matches = [index for index,item in enumerate(dsimm_arr) if item == 0]
#print(matches
#CHECKING IF VALID FOR 2 ENTITES :
#print(dsimm_arr[6])
mn_dsimm=np.mean(dsimm_arr)
print(f"Disimilarity: {mn_dsimm:.2f}")
# Similarity is 1-dissim
simm=1-mn_dsimm
print(f"Similarity: {simm:.2f}")
```
